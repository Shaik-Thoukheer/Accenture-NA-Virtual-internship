```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"D:\data analyst\projects & internships\0.  internships scn n enroll\2. Accenture\rw Content.csv")
pd.set_option('display.max.rows',1000) 
df
```

```python
# List of column names to be deleted
columns_to_delete = ['Unnamed: 0']  # Replace with the actual column names

# Drop the specified columns
df.drop(columns=columns_to_delete, inplace=True)
```

```python
print(df.dtypes)
```

```python
df['Type'] = pd.Categorical(df['Type'])
df['Category'] = pd.Categorical(df['Category'])
```

```python
df.info()
```

```python
ndc = non_duplicates_count = (~df.duplicated()).sum()

ndc
```

```python
df.isnull()
```

```python
df.dropna(inplace=True)
```

```python
from fuzzywuzzy import fuzz

# Defining a function to find the index of the most similar name in the 'Type' column
def find_most_similar_index_type(x):
    similarities = df['Type'].apply(lambda y: fuzz.ratio(x, y))
    return similarities.idxmax()

# Apply the function to each entry in the 'Type' column
similar_names_type = df['Type'].apply(find_most_similar_index_type)

# Defining a function to find the index of the most similar name in the 'Category' column
def find_most_similar_index_category(x):
    similarities = df['Category'].apply(lambda y: fuzz.ratio(x, y))
    return similarities.idxmax()

# Apply the function to each entry in the 'Category' column
similar_names_category = df['Category'].apply(find_most_similar_index_category)
```

```python
# Convert text data to lowercase
df['Type'] = df['Type'].str.lower()
df['Category'] = df['Category'].str.lower()

# Strip leading and trailing whitespaces
df['Type'] = df['Type'].str.strip()
df['Category'] = df['Category'].str.strip()
```

```python
plt.figure(figsize=(12, 6))  # Adjust figure size as needed

# Histogram for 'Type' column
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Type', hue='Type', palette='bright', multiple='stack')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend().remove()

# Histogram for 'Category' column
plt.subplot(1, 2, 2)
sns.histplot(data=df, x='Category', hue='Category', palette='bright', multiple='stack')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend().remove()

plt.tight_layout()  # Adjust subplot layout
plt.show()
```
