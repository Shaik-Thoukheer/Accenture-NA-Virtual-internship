```python
# Convert the data types of DataFrame columns
merged_df['Type_x'] = merged_df['Type_x'].astype('category')
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'])
merged_df['Sentiment'] = merged_df['Sentiment'].astype('category')
merged_df['Score'] = merged_df['Score'].astype('int64')
merged_df['User ID_y'] = merged_df['User ID_y'].astype('object')
merged_df['Type_y'] = merged_df['Type_y'].astype('category')
merged_df['Category'] = merged_df['Category'].astype('category')

# Save the DataFrame to a CSV file
merged_df.to_csv(r"D:\data analyst\projects & internships\0.  internships scn n enroll\2. Accenture\Merged.csv", index=False)
```

```python
# Rename columns 'Type_x' and 'Type_y'
merged_df = mdf.rename(columns={'Type_x': 'Reaction_type', 'Type_y': 'File_type'}, inplace=True)

drop_col = ['User ID_x','User ID_y']
mdf.drop(columns=drop_col, inplace=True)

mdf
```

```python
# Convert text data to lowercase
mdf['Reaction_type'] = mdf['Reaction_type'].str.lower()
mdf['File_type'] = mdf['File_type'].str.lower()
mdf['Category'] = mdf['Category'].str.lower()

# Strip leading and trailing whitespaces
mdf['Reaction_type'] = mdf['Reaction_type'].str.strip()
mdf['File_type'] = mdf['File_type'].str.strip()
mdf['Category'] = mdf['Category'].str.strip()
```

```python
plt.figure(figsize=(18, 6))  # Adjust figure size as needed

# Sort the DataFrame by the 'Reaction_type' column in ascending order
mdf_sorted = mdf.sort_values(by='Reaction_type', ascending=False)
# Histogram for 'Type' column
plt.subplot(1, 3, 1)
sns.histplot(data=mdf, x='Reaction_type', hue='Reaction_type', palette='bright', multiple='stack')
plt.xlabel('Reaction_type')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend().remove()

# Sort the DataFrame by the 'File_type' column in ascending order
mdf_sorted = mdf.sort_values(by='File_type', ascending=False)
# Histogram for 'Category' column
plt.subplot(1, 3, 2)
sns.histplot(data=mdf, x='File_type', hue='File_type', palette='bright', multiple='stack')
plt.xlabel('File_type')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend().remove()

# Sort the DataFrame by the 'Category' column in ascending order
mdf_sorted = mdf.sort_values(by='Category', ascending=False)

# Histogram for 'Category' column
plt.subplot(1, 3, 3)
sns.histplot(data=mdf, x='Category', hue='Category', palette='bright', multiple='stack')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend().remove()

plt.tight_layout()  # Adjust subplot layout
plt.show()
```
