```python
ndc = non_duplicates_count = (~df.duplicated()).sum()

ndc
```

```python
columns_to_delete = ['Unnamed: 0']
df.drop(columns=columns_to_delete, inplace = True )
```

```python
plt.figure(figsize=(12, 6))  # Adjust figure size as needed

# Histogram for 'Type' column
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Type', hue='Type', palette='bright', multiple='stack')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend().remove()

# Histogram for 'Sentiment' column
plt.subplot(1, 2, 2)
sns.histplot(data=df, x='Sentiment', hue='Sentiment', palette='bright', multiple='stack')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend().remove()

plt.tight_layout()  # Adjust subplot layout
plt.show()
```
